# Linear Regressions

### Necessary Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### Getting Data From Dataset

And checking a sample 

```python
co2_emmisions = pd.read_csv('./CO2_Emissions_Canada.csv')
co2_emmisions.head(5)
```

## About The CSV

This dataset contains official record of CO2 emissions data by various car of different features.

There are total 7385 rows and 12 columns. There are few abbreviations that has been used to describe the features. Those are listed in the Data Description sheet.

Further Info About The CSV:

```python
print(co2_emmisions.describe())
print('\n')
print(co2_emmisions.info())
```

### Setting the X and Y and reshape the X to fit the matrix

```python
X = co2_emmisions['Fuel Consumption Comb (L/100 km)'].values.reshape(-1,1)
Y = co2_emmisions['CO2 Emissions(g/km)'].values
```

### Prepareing Values and spliting between test and train values

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=2020)
```

### Visualizing The Data

```python
%matplotlib inline
plt.scatter(x_train, y_train, color='blue')
plt.xlabel('CO2 Emissions(g/km)')
plt.ylabel('Fuel Consumption Comb (L/100 km)')
plt.title('CO2 Emissions In Canada')
plt.show()
```

### Creating the LinearRegression and setting the test values

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Training model
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

### Accuracy Test

```python
print(f'Testing Accuracy: {model.score(x_test, y_test)*100}%')
print(f'Training Accuracy: {model.score(x_train, y_train)*100}%')
```

### Visualizing The Data with the LinearRegression

```python
%matplotlib inline
plt.scatter(x_train, y_train, color='blue')
plt.scatter(x_test, y_test, color='y')
plt.plot(x_test, y_predict, color='red')
plt.xlabel('CO2 Emissions(g/km)')
plt.ylabel('Fuel Consumption Comb (L/100 km)')
plt.title('CO2 Emissions In Canada')
plt.show()

```

## Conclusions:

- With the Increase of Fuel Consumption, CO2 Emissions Increases for sure(XD)
- This LinearRegression could be used to insure that if newly produced cars is better in ecological terms
- The CO2 Emissions for Cars with the same fuel almost equal and the diffrance is the Fuel Consumption only
